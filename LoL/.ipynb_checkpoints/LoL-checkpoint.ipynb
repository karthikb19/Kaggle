```python
clean_df = df.copy()
dropcols = ['blueWardsPlaced', 'blueWardsDestroyed', 'redWardsPlaced', 'redWardsDestroyed', 'blueFirstBlood',
           'redFirstBlood', 'redHeralds', 'blueHeralds', 'blueEliteMonsters', 'redEliteMonsters', 'blueTotalMinionsKilled', 'redTotalMinionsKilled', 'blueTotalJungleMinionsKilled', 'redTotalJungleMinionsKilled', 'redTotalExperience', 'redTotalGold', 'redKills',

 'redDragons',
 'redAvgLevel',
 'redGoldDiff',
 'redExperienceDiff',
 'redCSPerMin',
 'redGoldPerMin', 'blueExperienceDiff']

clean_df = clean_df.drop(dropcols, axis=1)
```

```python
clean_df.head()
```

```python
correlations = clean_df[clean_df.columns[1:]].apply(lambda x: x.corr(clean_df['blueWins']))
cols = []
for col in correlations.index:
    if (correlations[col]>0.2 or correlations[col]<-0.2):
        cols.append(col)
cols
```

```python
clean_df = clean_df[cols]
clean_df.head()
```

# Final Clean Up

```python
from sklearn.preprocessing import MinMaxScaler
X = clean_df
y = df["blueWins"]

scaler = MinMaxScaler()
scaler.fit(X)
X = scaler.transform(X)


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

# Ensemble Learning

```python
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.linear_model import LogisticRegression, SGDClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

lr = LogisticRegression(random_state=42)
rf = RandomForestClassifier(n_estimators=100, max_depth=6, random_state=42)
nb = GaussianNB()
svc = SVC(C=0.98, gamma="scale", probability=True, random_state=42)
sgd = SGDClassifier(random_state=42)
gb = GradientBoostingClassifier(random_state=42)
knn = KNeighborsClassifier(n_neighbors=7)

# labels = ['LogisticReg', 'RandomForest', 'Naive Bayes', 'SVM', 'SGD', 'GradientBoost', 'KNN']


# for clf, label in zip([lr, rf, nb, svc, sgd, gb, knn], labels):
#     scores = model_selection.cross_val_score(clf, X_train, y_train, cv=5, scoring='accuracy')
#     print('Accuracy: %0.2f (+/- %0.2f) [%s]' % (scores.mean(), scores.std(), label))
```

```python
from sklearn.metrics import accuracy_score
lr.fit(X_train, y_train)
rf.fit(X_train, y_train)
nb.fit(X_train, y_train)
svc.fit(X_train, y_train)
sgd.fit(X_train, y_train)
gb.fit(X_train, y_train)
knn.fit(X_train, y_train)


lr_pred = lr.predict(X_test)
rf_pred = rf.predict(X_test)
nb_pred = nb.predict(X_test)
svc_pred = svc.predict(X_test)
sgd_pred = sgd.predict(X_test)
gb_pred = gb.predict(X_test)
knn_pred = knn.predict(X_test)

print("LR: ", accuracy_score(lr_pred, y_test))
print("RF: ", accuracy_score(rf_pred, y_test))
print("NB: ", accuracy_score(nb_pred, y_test))
print("SVC: ", accuracy_score(svc_pred, y_test))
print("SGD: ", accuracy_score(sgd_pred, y_test))
print("GB: ", accuracy_score(gb_pred, y_test))
print("KNN: ", accuracy_score(knn_pred, y_test))
```

```python
from sklearn.ensemble import VotingClassifier
voting_clf = VotingClassifier(estimators=[('lr', lr), ('rf', rf), ('nb', nb), ('svc', svc), ('sgd', sgd), ('gb', gb), ('knn', knn)], voting='hard')

voting_clf.fit(X_train, y_train)
preds = voting_clf.predict(X_train)
acc = accuracy_score(y_train, preds)
print('Acc: ', acc)
```
